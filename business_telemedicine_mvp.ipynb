# Expanding Telehealth to Offset Costs of Missed Appointments 

## **Business Opportunity**
The COVID-19 pandemic and Medicare's temporary [expansion of covered telehealth services](https://www.medicare.gov/coverage/telehealth) accelerated the adoption of patients meeting with their doctor via phone and/or video. While this was crucial during the pandemic to help people access care without needing to take public transportation and congregate in medical waiting rooms, it may have opened an opportunity for medical practices to avoid the costs associated to patients missing their medical appointments. Unused 60-minute time-slots cost medical practices [`$` 200 on average](https://www.hcinnovationgroup.com/clinical-it/article/13008175/missed-appointments-cost-the-us-healthcare-system-150b-each-year), costing the U.S. `$` 150 billion each year from missed appointments alone.

## **Solution Path**
Exploratory data analysis to see how many Medicare patients over the age of 75 in urban areas used telehealth services during 2021 to inform decision about investing in the technology. 

## **Impact Hypothesis**
Offering telehealth services to patients would reduce costs the medical practice incurrs due to no-shows.

The Centers for Medicare & Medicaid Services publishes the data in a [Public Use File](https://www.cms.gov/Research-Statistics-Data-and-Systems/Downloadable-Public-Use-Files/MCBS-Public-Use-File). 

The Winter 2021 survey asked participants about their use of telehealth services (video and audio), their access to technology (internet at home, computers, smartphones, tablets), and their use of video conferencing services (e.g., Zoom, Skype, FaceTime).
